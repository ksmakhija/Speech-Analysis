### Stopword removal 

```python
stop_words = set(stopwords.words('english') + list(string.punctuation) + ['--', '``', "''","'"])
```

```python
clean_roosevelt=[]
```

```python
for x in roosevelt_speech.split():
    if x not in stop_words:
        x=x.lower()
        x=x.replace("\'", "")
        clean_roosevelt.append(x) 
```

```python
 roosevelt_str = ' '.join(clean_roosevelt)
```

```python
roosevelt_str
```

```python
clean_kennedy=[]
```

```python
for x in kennedy_speech.split():
    if x not in stop_words:
        x=x.lower()
        x=x.replace("\'", "")
        clean_kennedy.append(x)
```

```python
 kennedy_str = ' '.join(clean_kennedy)
```

```python
kennedy_str
```

```python
for x in nixon_speech.split():
    if x not in stop_words:
        x=x.lower()
        x=x.replace("\'", "")
        clean_nixon.append(x) 
```
